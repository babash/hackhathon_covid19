## Anxiété

```python
anxiety_df = pd.read_csv('Documents/anxiety_tweets.csv')
anxiety_df['date']= pd.to_datetime(anxiety_df['date'])
```

```python
output_df = pd.DataFrame(anxiety_df.set_index('date').resample('W-MON').count()['content'])
```

```python
output_df.reset_index(inplace=True)
output_df = output_df.drop([output_df.index[120]])

output_df = output_df.rename(columns={'content': 'anxiety_tweets_counts'})
#output_df
```

```python
confinement = []
for i in range(120):
    if (i >=62 and i <=70) or (i >=94 and i <=101):
        confinement.append("oui")
    else:
        confinement.append("non")
output_df['confinement'] = pd.Series(confinement)
#62 à 70 confinement 1 
#94 à 101 confinement 2
#102 à ajd couvre feu 
```

```python
import plotly.express as px
fig = px.bar(output_df, x='date', y='anxiety_tweets_counts', color="confinement", title="Weekly counts of french language tweets with the word 'anxiété' between 2019 and today", labels={"anxiety_tweets_counts":"Number of tweets", "date":"Weeks"})
fig.show()
```

```python
import plotly.express as px

#df = px.data.tips()
fig = px.scatter(output_df, x="date", y="anxiety_tweets_counts", trendline="lowess", title="Scatterplot of counts of french language tweets with the word 'anxiété' between 2019 and today with Trendline", labels={"anxiety_tweets_counts":"Number of tweets", "date":"Weeks"})
fig.show()
```

```python
#Independent Student T-Test to see if there is a significant difference in the number of tweets between 2020 and 2021 
from scipy.stats import ttest_rel
stat, p = ttest_rel(output_df['anxiety_tweets_counts'][0:51], output_df['anxiety_tweets_counts'][52:103])
if p < 0.05:
    print("There is a significant difference between the number of tweets related to 'anxiety' in 2019 and 2020. p-value: ", p)
else: 
     print("There is no significant difference between the number of tweets related to 'anxiety' in 2019 and 2020. p-value: ", p)
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Bar(
    x=['2019', '2020'], y = [output_df['content'][0:51].mean(), output_df['content'][52:103].mean()],
    error_y=dict(type='data', array=[output_df['content'][0:51].std(), output_df['content'][52:103].std()])
))
"""fig.add_trace(
    go.Scatter(
        x=["2019", "2020"],
        y=[3000,3000],   
        mode="lines+text",
        name="Lines Text",
        text=["***" ],
        textposition="top center"
    )
)"""

fig.update_layout(barmode='group')
fig.show()
```
